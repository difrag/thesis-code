```python
stages = adata.obs["braak_stage"].astype(str).unique()
for stage in stages:
    # subset
    a_sub = adata[adata.obs["braak_stage"].astype(str) == stage].copy()

    # ensure raw counts in .X
    if a_sub.raw is not None:
        a_sub.X = a_sub.raw.X.copy()

    # counts file
    counts_fn = f"counts_Braak{stage}.h5ad"
    a_sub.var_names = a_sub.var["gene_ids"]
    a_sub.write_h5ad(counts_fn, compression="gzip")

    # meta file
    meta = (
        a_sub.obs[["cell_type"]]
             .reset_index()
             .rename(columns={"index": "Cell"})
    )
    meta_fn = f"meta_Braak{stage}.txt"
    meta.to_csv(meta_fn, sep="\t", index=False)

    print(f"Saved {counts_fn} & {meta_fn}  ({a_sub.n_obs:,} cells)")
```

```python
# 0. PREP ─────────────────────────────────────────────────────────────
import scanpy as sc
import pandas as pd
import re

# adata = sc.read_h5ad("your_original.h5ad")   # ← uncomment & edit

# --------------------------------------------------------------------
# 1. DROP the unwanted cell type(s)
# --------------------------------------------------------------------
# Anything matching either spelling (with / without dash / stray spaces)
pattern = re.compile(r"Brain\s*region[-\s]+specific\s+neurons", re.IGNORECASE)

mask_keep = ~adata.obs["cell_type"].astype(str).str.match(pattern)
print(f"Keeping {mask_keep.sum():,} cells out of {adata.n_obs:,}")

adata = adata[mask_keep].copy()        # filtered AnnData

# --------------------------------------------------------------------
# 2. COUNTS  (CellphoneDB wants raw counts with gene IDs as var_names)
# --------------------------------------------------------------------
if adata.raw is not None:
    adata.X = adata.raw.X.copy()

adata.var_names = adata.var["gene_ids"]

adata.write_h5ad("cpdb_counts.h5ad", compression="gzip")
# sc.write_10x_mtx("cpdb_counts_10x", adata,
#                  gene_symbols=adata.var["gene_symbol"], overwrite=True)

# --------------------------------------------------------------------
# 3. META  (Cell  |  cell_type)
# --------------------------------------------------------------------
meta = (
    adata.obs[["cell_type"]]
         .reset_index()
         .rename(columns={"index": "Cell"})
)
meta.to_csv("meta_cpdb.txt", sep="\t", index=False)

# --------------------------------------------------------------------
# 4. SANITY CHECK
# --------------------------------------------------------------------
bad_names = (
    meta["cell_type"].str.contains("-") |
    meta["cell_type"].str.match(r"^\d+$")
)
print(f"{bad_names.sum()} problematic cell-type names (should be 0)")
```
